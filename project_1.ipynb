```python
import pandas as pd
ds_df=pd.read_csv(r'/content/drive/My Drive/Diwali Sales Data.csv',encoding='latin')
```

```python
display(ds_df)
```

# **1. How many males and females plot the count of gender**

```python
import plotly.express as px
fig=px.histogram(ds_df,x='Gender',title='count of Gender',color='Gender')
fig.show()
```

# **2. For male and female Give me the sum of amt wrt each gender and plot the graph**

```python
ds_df.groupby('Gender')['Amount'].sum()
fig=px.histogram(ds_df,x='Gender',y='Amount',title='Total Amount by Gender',color='Gender')
fig.show()
```

# **3. Count of each age group with individual count of age group wrt gender**

```python
import matplotlib.pyplot as plt
age_gender_counts=ds_df.groupby(['Age Group','Gender'])['Gender'].count()
age_gender_counts.plot.bar(color=['blue','red','green','orange','black','pink'])
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Count of each age group by gender')
plt.show()
```

# **4. Plot the total amt spent by each age group**


```python
ds_df.groupby('Age Group')['Amount'].sum ()
fig=px.histogram(ds_df,x='Age Group',y='Amount',title='Total Amount by Age Group',color='Age Group')
fig.show()
```

# **5. Plot the graph that depicts total number of orders from top 10 states**

```python
ds_df.groupby('State')['Orders'].sum().sort_values(ascending=False).head(10)
ds_df.groupby('State')['Orders'].sum().sort_values(ascending=False).head(10)
fig = px.bar(ds_df.groupby('State')['Orders'].sum().sort_values(ascending=False).head(10),
             x=ds_df.groupby('State')['Orders'].sum().sort_values(ascending=False).head(10).index,
             y=ds_df.groupby('State')['Orders'].sum().sort_values(ascending=False).head(10).values,
             title='Total Number of Orders from Top 10 States',
             color = ds_df.groupby('State')['Orders'].sum().sort_values(ascending=False).head(10).index)
fig.show()
```

# **6. Total amt of top 10 states**

```python
state_amount=ds_df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10)
fig = px.bar(state_amount,
             x=state_amount.index,
             y=state_amount.values,
             title='Total Amount Spent by Top 10 States',
             color = state_amount.index)
fig.show()
```

# **7. Plot me a graph that shows comparison between no of married and unmarried people**

# **8. Plot me amt spent by each male and each female wrt to marital status**

# **9. Plot me the count of each occupation that is present in data set**

```python
fig=px.histogram(ds_df,x='Occupation',title='count of Occupation',color='Occupation')
fig.show()
```

# **10. Plot the amt spent by each occupation in descending order**

```python
ds_df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False)
temp=ds_df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).index
fig = px.bar(ds_df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False),
             x=temp,
             y=ds_df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).values,
             title='Amount spent by each occupation',
             color=temp)
fig.show()
```

# **11. Plot the graph provides statistics on the interest of customers wrt to product category**

```python
fig = px.histogram(ds_df, x='Product_Category', title='Number of Purchases per Product Category',color='Product_Category')
fig.show()
```

# **12. I would like to know the budget spent on each product category in descending order**


```python
ds_df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False)
fig = px.bar(ds_df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(10),
             x=ds_df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(10).index,
             y=ds_df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(10).values,
             title='Total Amount Spent by Top 10 States',
             color = ds_df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(10).index)
fig.show()
```

# **13. Provide me statical analysis of each product based upon the percentage of orders that has been completed**

```python
product_analysis=ds_df.groupby('Product_Category')['Orders'].sum()/ds_df['Orders'].sum()*100
product_analysis_df = product_analysis.reset_index(name='Percentage of Orders')
fig = px.bar(product_analysis_df, x='Product_Category', y='Percentage of Orders',
             title='Percentage of Orders for Each Product',
             color='Product_Category',
             labels={'Product_Category': 'Product Category', 'Percentage of Orders': 'Orders (%)'})
fig.show()
```
